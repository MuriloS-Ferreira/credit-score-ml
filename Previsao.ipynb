```python
import pandas as pd

tabela = pd.read_csv("data/clientes.csv")
display(tabela)
```

```python
#como a IA só consegue trabalhar com números, precisamos transformar os textos em números
#antes vamos ver quais colunas possuem textos
display(tabela.info())
from sklearn.preprocessing import LabelEncoder
#codificar as colunas
codificador_profissao = LabelEncoder()
codificador_mixcredito = LabelEncoder()
codificador_comportamento = LabelEncoder()

#aplicar a codificação
tabela["profissao"] = codificador_profissao.fit_transform(tabela["profissao"])
tabela["mix_credito"] = codificador_mixcredito.fit_transform(tabela["mix_credito"])
tabela["comportamento_pagamento"] = codificador_comportamento.fit_transform(tabela["comportamento_pagamento"])
display(tabela)
```

```python
#separar os dados em qual vamos usar para prever e quais são os dados que vamos usar para fazer a previsão, esses dados são separados em x e y
#tirar a coluna que queremos prever de dentro da tabela, o x é tudo menos a coluna que queremos prever
x=tabela.drop(columns=["score_credito", "id_cliente"])
#ja o y é a coluna que queremos prever
y=tabela["score_credito"]
from sklearn.model_selection import train_test_split
#separar os dados em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
#treinar o modelo
modelo_arvore = RandomForestClassifier()
modelo_arvore.fit(x_treino, y_treino)
#avaliar a precisão do modelo
from sklearn.metrics import accuracy_score
previsoes = modelo_arvore.predict(x_teste)
#mostra como foi a precisão do modelo
display (accuracy_score(y_teste, previsoes))
```

```python
novos_clientes = pd.read_csv("data/novos_clientes.csv")
display(novos_clientes)
display(novos_clientes.info())
#aplicar a codificação nos novos clientes
novos_clientes["profissao"] = codificador_profissao.transform(novos_clientes["profissao"])
novos_clientes["mix_credito"] = codificador_mixcredito.transform(novos_clientes["mix_credito"])
novos_clientes["comportamento_pagamento"] = codificador_comportamento.transform(novos_clientes["comportamento_pagamento"])
display(novos_clientes)
#fazer a previsão dos novos clientes
previsoes_novos = modelo_arvore.predict(novos_clientes)
display(previsoes_novos)
```
